## Remove Stop words

```python
import nltk

# Download necessary NLTK data files
nltk.download('stopwords')
#Downloads the NLTK stopwords dataset (common words like "the," "is," etc.).
nltk.download('punkt')

from nltk.corpus import stopwords
#Imports the stopwords module from NLTK's corpus for accessing stop words.

# Input text
text = """Sri Lanka is an island nation located in the Indian Ocean, just off the southeastern coast of India.
Despite its modest size—slightly larger than the state of West Virginia—Sri Lanka has a population of about 20 million people,
almost equal to the population of Texas. The island is rich in natural resources, and has a diverse economy based on agriculture,
mining, fishing, manufacturing, and tourism. On becoming an independent nation in 1948 Sri Lanka"""

# Define stop words
stop_words_set = set(stopwords.words('english'))
#Creates a set of English stop words (e.g., "is," "an") for efficient lookup.

# Tokenize text
tokenized_words = nltk.word_tokenize(text)
#Splits the input text into individual words/tokens using NLTK's tokenizer.

# Remove stop words
tokenized_words_no_stop_words = [word for word in tokenized_words if word.lower() not in stop_words_set]
#Creates a list of tokens, excluding stop words (case-insensitive comparison).

# Outputs
#print("Stop Words List:", stop_words_set)  # Show the list of stop words
print("\nWords Removed (Stop Words):", set(tokenized_words) - set(tokenized_words_no_stop_words))
#Prints the stop words that were removed by comparing original and filtered token sets.
print("\nOriginal Tokens:", tokenized_words)  # Show original tokenized words
print("\nTokens without Stop Words:", tokenized_words_no_stop_words)  # Show cleaned tokens
```

#  Stemming and Lemmatization

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
#Imports PorterStemmer for stemming and WordNetLemmatizer for lemmatization.

from nltk.tokenize import word_tokenize
#Imports word_tokenize to split text into individual words.

from nltk.corpus import wordnet
#Imports WordNet corpus for lemmatization, providing lexical data.

# Download necessary NLTK data files
nltk.download('punkt')
#Downloads the 'punkt' tokenizer for word tokenization.
nltk.download('wordnet')
#Downloads WordNet for lemmatization, a lexical database of English words.
nltk.download('omw-1.4')  # For better lemmatizer results
#Downloads Open Multilingual WordNet for enhanced lemmatization support.

# Input text
text = """Sri Lanka is an island nation located in the Indian Ocean.
Despite its modest size, it has a diverse economy based on agriculture, mining, and tourism."""

# Tokenize the text
tokenized_words = word_tokenize(text)
#Splits the text into individual words (tokens).

# Initialize the Porter Stemmer and WordNet Lemmatizer
stemmer = PorterStemmer()
#Initializes the Porter Stemmer for reducing words to their root form.
lemmatizer = WordNetLemmatizer()
#Initializes the WordNet Lemmatizer for reducing words to their base form using WordNet.

# Perform Stemming
stemmed_words = [stemmer.stem(word) for word in tokenized_words]
#Applies stemming to each tokenized word, reducing it to its root (e.g., "running" → "run").

# Perform Lemmatization (default POS is 'n' for nouns)
lemmatized_words = [lemmatizer.lemmatize(word) for word in tokenized_words]
#Applies lemmatization to each tokenized word, reducing it to its base form (e.g., "running" → "run") with default part-of-speech (POS) as noun.

# Display Results
print("Original Words:\n", tokenized_words)
print("\nStemmed Words:\n", stemmed_words)
print("\nLemmatized Words (default POS=noun):\n", lemmatized_words)

```

# Text normalization: Lowercasing, removing special characters, handling contractions.

```python
import re
#Imports the re module for regular expression operations.
import string
#Imports the string module for string-related utilities

# Define a contraction mapping
CONTRACTIONS_MAP = {
    "can't": "cannot",
    "won't": "will not",
    "isn't": "is not",
   # "Isn't": "is not",
    "aren't": "are not",
    "wasn't": "was not",
    "weren't": "were not",
    "haven't": "have not",
    "hasn't": "has not",
    "hadn't": "had not",
    "doesn't": "does not",
    "don't": "do not",
    "didn't": "did not",
    "couldn't": "could not",
    "shouldn't": "should not",
    "wouldn't": "would not",
    "mustn't": "must not",
    "it's": "it is",
    "he's": "he is",
    "she's": "she is",
    "that's": "that is",
    "there's": "there is",
    "what's": "what is",
    "who's": "who is",
    "let's": "let us",
    "i'm": "i am",
    "you're": "you are",
    "they're": "they are",
    "we're": "we are",
    "i've": "i have",
    "you've": "you have",
    "we've": "we have",
    "they've": "they have",
    "i'll": "i will",
    "you'll": "you will",
    "he'll": "he will",
    "she'll": "she will",
    "we'll": "we will",
    "they'll": "they will",
    "isn't": "is not",
    "wasn't": "was not",
    "weren't": "were not",
    "there's": "there is",
}
#Defines a dictionary mapping contractions (e.g., "can't") to their expanded forms (e.g., "cannot").

def expand_contractions(text, contractions_map):
    """Expand contractions in a given text using a mapping dictionary."""
    for contraction, expansion in contractions_map.items():
        #Iterates over each contraction-expansion pair in contractions_map.
        text = re.sub(r"\b" + re.escape(contraction) + r"\b", expansion, text)
        #Replaces each contraction with its expansion, ensuring word boundaries (\b) to avoid partial matches.
    return text

#Defines a function to normalize text for NLP preprocessing.
def normalize_text(text):
    """Normalize text: Lowercasing, removing special characters, and handling contractions."""
    # Step 1: Expand contractions
    #Expands contractions in the input text using the expand_contractions function.
    text = expand_contractions(text, CONTRACTIONS_MAP)

    # Step 2: Lowercasing
    #Converts all characters in text to lowercase.
    text = text.lower()

    # Step 3: Remove special characters (retain only letters and spaces)
    #Removes all characters except lowercase letters and spaces using a regular expression.
    text = re.sub(r"[^a-z\s]", "", text)

    # Step 4: Remove extra spaces
    #Replaces multiple spaces with a single space and removes leading/trailing spaces.
    text = re.sub(r"\s+", " ", text).strip()

    return text

# Example text
text = """I can't believe it's already 2023! Isn't this amazing? Let's clean up this text for NLP preprocessing."""

# Normalize the text
normalized_text = normalize_text(text)

# Output results
print("Original Text:\n", text)
print("\nNormalized Text:\n", normalized_text)
```

# Bag Of Word (BoW) Representation

```python
from sklearn.feature_extraction.text import CountVectorizer
#Imports the CountVectorizer class from scikit-learn to convert text into a numerical Bag of Words representation.

# Example documents go here.
documents = [
    "I love Galle.",
    "Galle is fun.",
    "I love fun activities."
]

# Create an instance of CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the documents into a Bag of Words representation
bow_matrix = vectorizer.fit_transform(documents)
#Fits the vectorizer to the documents and transforms them into a sparse Bag of Words matrix, counting word occurrences.

# Get the vocabulary (unique words)
vocabulary = vectorizer.get_feature_names_out()
#Retrieves the unique words (features) from the documents as the vocabulary.

# Convert the BoW matrix to a dense array for better readability
bow_dense = bow_matrix.toarray()
#Prints the list of unique words in the vocabulary.

# Display the results
print("Vocabulary:", vocabulary)
print("\nBag of Words Representation (Dense Matrix):")
for i, vector in enumerate(bow_dense):
    print(f"Document {i+1}: {vector}")
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#Imports the TfidfVectorizer class from scikit-learn to convert text into TF-IDF features.
#TF-IDF (Term Frequency-Inverse Document Frequency) is a numerical statistic used in text mining and information retrieval
# to measure the importance of a word in a document relative to a collection of documents (corpus).

#Term Frequency (TF): Measures how frequently a word appears in a document.
#It's calculated as the number of times a word appears divided by the total number of words in the document.
#Inverse Document Frequency (IDF): Measures how unique or rare a word is across the entire corpus.
#It's calculated as the logarithm of the total number of documents divided by the number of documents containing the word.
#TF-IDF Score: The product of TF and IDF. A high score indicates a word is frequent in a document but rare across the corpus,
# making it significant for that document.



# Example documents
documents = [
    "I love programming.",
    "Programming is bad.",
    "I love fun activities activities."
]

# Create TF-IDF model
vectorizer = TfidfVectorizer()
#Creates an instance of the TfidfVectorizer to build a TF-IDF model.
tfidf_matrix = vectorizer.fit_transform(documents)
#Fits the vectorizer to the documents and transforms them into a TF-IDF matrix.

# Display vocabulary and TF-IDF matrix
print("Vocabulary:", vectorizer.get_feature_names_out())
#Prints the unique words (features) extracted from the documents.
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
#Prints the TF-IDF matrix as a dense array, showing the TF-IDF scores for each word in each document.

from transformers import pipeline

# Load lightweight DistilBERT sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Analyze sentiment for each document
for i, doc in enumerate(documents):
    result = sentiment_analyzer(doc)[0]
    print(f"Document {i+1}: '{doc}' -> Sentiment: {result['label']} (Confidence: {result['score']:.4f})")
```
